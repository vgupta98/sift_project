```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
img = cv2.imread('grp_img.jpg')
```

```python
plt.imshow(img, cmap='gray')
```

```python
sift=cv2.xfeatures2d.SIFT_create(nOctaveLayers=5,contrastThreshold=0.08,edgeThreshold=5)
```

```python
kp,desc=sift.detectAndCompute(img,None)
```

```python
print(desc.shape)
```

```python
print(len(kp))
```

```python
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
img=cv2.drawKeypoints(gray,kp,outImage=None,flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

cv2.imwrite('sift_keypoints.jpg',img)
```

```python
keypoint_img=cv2.imread('sift_keypoints.jpg')
```

```python
cv2.imshow('image',keypoint_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
haar_cascade=cv2.CascadeClassifier('F:\\opencv3.3.1\\build\\etc\\haarcascades\\haarcascade_frontalface_alt.xml')
```

```python
#let's detect multiscale (some images may be closer to camera than others) images 
faces = haar_cascade.detectMultiScale(gray2, scaleFactor=1.2, minNeighbors=5);  
 
#print the number of faces found 
print ('Faces found: ', len(faces))
```

```python
for (x, y, w, h) in faces:     
    cv2.rectangle(gray2, (x, y), (x+w, y+h), (0, 255, 0), 2)
```

```python
plt.imshow(gray2, cmap='gray');
```

```python
gray3=cv2.imread('grp_img.jpg',0)
```

```python
thresh=cv2.adaptiveThreshold(gray3,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,\
            cv2.THRESH_BINARY,11,2)
```

```python
cv2.imshow('image',thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(thresh,cmap='gray')
```

```python
sift3=cv2.xfeatures2d.SIFT_create(nOctaveLayers=5,contrastThreshold=0.08,edgeThreshold=5)
```

```python
kp3=sift3.detect(thresh,None)
```

```python
thresh=cv2.drawKeypoints(gray3,kp3,outImage=None,flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

cv2.imwrite('sift_keypoints3.jpg',thresh)
```

```python
keypoint_img3=cv2.imread('sift_keypoints3.jpg')
```

```python
cv2.imshow('image',keypoint_img3)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
